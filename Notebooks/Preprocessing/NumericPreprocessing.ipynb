```python
import numpy as np
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/ML_Results_Main/original.csv")

df.head(5)
```

```python
print(df.shape)
df.describe()
```

```python
# Check for null age
df.isnull().sum()
```

```python
# Drop null age
df = df.dropna(subset = ['age'])
```

```python
# Check for negative age
df[df['age'] < 0]
```

```python
# Drop negative age
negativeAge = df[df['age'] < 0].index
df.drop(negativeAge, axis=0, inplace=True)
```

```python
# Check again for negative age
df[df['age'] < 0]
```

```python
# Drop clientid and shuffle rows
df = df.drop(columns=['clientid'])
df = df.sample(frac=1).reset_index(drop=True)
df
```

```python
df.to_csv("/content/drive/MyDrive/ML_Results/preprocessed.csv", index=False, index_label=True)
```
